```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Bookings-100000-Rows.xlsx")

print("Shape:", df.shape)
df.head()
```

```python

nulls = df.isnull().sum()
nulls = nulls[nulls > 0]
print(nulls.sort_values(ascending=False))
```

```python
print(df['Time'].head(20))
print(df['Time'].dtype)
```

```python
df['Booking_Status'].value_counts(dropna=False)
```

```python
df['Incomplete_Rides_Reason'] = (
    df['Incomplete_Rides_Reason']
    .fillna('Ride Completed')
)
```

```python
df['Canceled_Rides_by_Customer'] = (
    df['Canceled_Rides_by_Customer']
    .fillna('Not Cancelled')
)
```

```python
df['Canceled_Rides_by_Driver'] = (
    df['Canceled_Rides_by_Driver']
    .fillna('Not Cancelled')
)
```

```python
df['Incomplete_Rides'] = (
    df['Incomplete_Rides']
    .fillna('No')
)
```

```python
df['Driver_Ratings'] = (
    df['Driver_Ratings']
    .fillna(0)
)

df['Customer_Rating'] = (
    df['Customer_Rating']
    .fillna(0)
)
```

```python
df['Payment_Method'] = (
    df['Payment_Method']
    .fillna('Not Applicable')
)
```

```python
print(df['V_TAT'].dtype)
print(df['C_TAT'].dtype)
```

```python
df['V_TAT'] = df['V_TAT'].fillna(0)
df['C_TAT'] = df['C_TAT'].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop(
    columns=['Vehicle Images'],
    inplace=True
)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Weekday'] = df['Date'].dt.day_name()

df['Hour'] = pd.to_datetime(
    df['Time'].astype(str),
    format='%H:%M:%S'
).dt.hour
```

```python
df.info()
```

```python
df['Booking_Status'].value_counts()
```

```python
# Revenue per KM
df['Revenue_per_KM'] = (
    df['Booking_Value'] /
    df['Ride_Distance']
)

# Peak Hours
df['Peak_Hour'] = np.where(
    ((df['Hour'] >= 7) & (df['Hour'] <= 10)) |
    ((df['Hour'] >= 17) & (df['Hour'] <= 21)),
    'Peak',
    'Non-Peak'
)

# Weekend Flag
df['Is_Weekend'] = np.where(
    df['Weekday'].isin(['Saturday','Sunday']),
    'Weekend',
    'Weekday'
)

# Rating Bucket
df['Driver_Rating_Group'] = pd.cut(
    df['Driver_Ratings'],
    bins=[-1,2,3,4,5],
    labels=['Poor','Average','Good','Excellent']
)
```
